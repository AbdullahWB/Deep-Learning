```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
n_pointer = 50
```

```python
class_0 = np.random.randn(n_pointer, 2) * 0.7 + np.array([-2, -2])
class_0
```

```python
class_1 = np.random.randn(n_pointer, 2) * 0.7 + np.array([2, 2])
class_1
```

```python
X = np.vstack((class_0, class_1))
X
```

```python
y = np.array([0]*n_pointer + [1]*n_pointer)
y
```

```python
df = pd.DataFrame(X, columns=["x1", "x2"])
df["target"] = y
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
plt.figure(figsize=(6, 5))
plt.scatter(df[df["target"]==0]["x1"],
           df[df["target"]==0]["x2"],
            label="Class 0")
plt.scatter(df[df["target"]==1]["x1"],
           df[df["target"]==1]["x2"],
            label="Class 1")
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Synthetic Linearly Separable Dataset")
plt.grid(True)
```
