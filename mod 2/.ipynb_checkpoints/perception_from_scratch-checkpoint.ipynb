```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
class Perceptron:
    def __init__(self, eta, epochs):
        self.weights = np.random.rand(3) * 1e-4
        print(f"Initial weights: {self.weights}")
        self.eta = eta
        self.epochs = epochs
    
    def activation_function(self, inputs, weights):
        z = np.dot(inputs, weights)
        return np.where(z > 0, 1, 0)
    
    def fit(self, X, y):
        self.X = X
        self.y = y
        
        x_with_bias = np.c_[self.X, -np.ones((len(self.X), 1))]
        print(f"X with bias:\n{x_with_bias}")
        
        for epoch in range(self.epochs):
            print(f"Epoch {epoch + 1}/{self.epochs}")
            y_hat = self.activation_function(x_with_bias, self.weights)
            print(f"Predictions for epoch {epoch + 1}: {y_hat}")
            
            error = self.y - y_hat
            print(f"Error for epoch {epoch + 1}: {error}")
            
            self.weights = self.weights + self.eta * np.dot(x_with_bias.T, error)
            print(f"Updated weights for epoch {epoch + 1}: {self.weights}")
            
            print("#"*20)
    
    def predict(self, X):
        x_with_bias = np.c_[X, -np.ones((len(X), 1))]
        return self.activation_function(x_with_bias, self.weights)
```

```python
import pandas as pd

data = {"x1": [0, 0, 1, 1], "x2": [0, 1, 0, 1], "y": [0, 0, 0, 1]}
and_gate = pd.DataFrame(data)

X = and_gate.drop("y", axis=1)
y = and_gate["y"]
y.to_frame()
```

```python
model = Perceptron(eta=0.5, epochs=10)
model.fit(X, y)
```
